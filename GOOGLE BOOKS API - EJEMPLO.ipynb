```python
#abrimos la librerias que necesitamos
import requests
import json
```

```python
#Obtenemos la api quedeseamos consultar, en este caso es google apis
result = requests.get("https://www.googleapis.com/books/v1/volumes?q=el+caballero+carmelo")
print(result.text) #Imprimimos la que contiene esta para poder analizarla
```

```python
book = result.json() #los resultados que se obtuvieron lo transformamos en json para trabajar de una manera más amigable
print(book.keys()) # Esto es para saber cuales son los elemntos principales
#Luego de esto analizamos cual es el elemento que tiene información más interesante y que deseamos obtener
#En nuestro caso la variable que presenta más información es: "items"
```

```python
items = book["items"] #obtenemos la información de items
print(items)#Imprimimos ...
```

```python
#Vamos a encodificar y luego decodificar debido a que hay información dentro de otra.
encoded = json.dumps(items)
decoded = json.loads(encoded)
print(decoded[0]) #Decoded es una lista y esta presenta solo 10 resultados(que son de la primera pestaña), por lo tanto
#Nos arrojara cada resultado especifico que querramos 
```

```python
#Como nos podemos percatar, hay llaves dentro de la lista que aún tienen más información y podemos acceder a ellas de la
#siguiente manera...
print(decoded[0]["volumeInfo"]) #Con esto podemos ver que es l que hay dentro de "volumeInfo"
```

```python
#Como podemos apreciar hay más llaves dentro de esta llave y lo unico que tenemos que hacer es seguir agregando corchetes y
#y dentro de estos colocar la información que requerimos
#Pero en esta ocasión nosotros deseamos extarer el resumen que esta con el nombre "description"
print(decoded[0]["volumeInfo"]["description"])
```

```python
# En este caso se elegio saber quien es el autor del libro escogido
print(decoded[0]["volumeInfo"]["authors"])
```
